```python
playerScores = [9.5, 8.9, 9.2, 9.8, 8.8, 9.0]
print('playerScores: {}'.format(playerScores))

minScore = 0
maxScore = 0
minScoreIdx = 0;
maxScoreIdx = 0

for idx, score in enumerate(playerScores):
    if idx == 0 or minScore > score:
        minScoreIdx = idx
        minScore = score
        
print('minScore: {}, minScoreIdx: {}'.format(minScore,minScoreIdx))
playerScores.pop(minScoreIdx)


playerScores = [9.5, 8.9, 9.2, 9.8, 8.8, 9.0]
for idx, score in enumerate(playerScores):
    if maxScore < score:
        maxScoreIdx = idx
        maxScore = score
        
print('maxScore: {}, maxScoreIdx: {}'.format(maxScore, maxScoreIdx))
playerScores.pop(maxScoreIdx)

print('playerScores: {}'.format(playerScores))
```

```python
#remove()
students = ['홍길동', '박찬호', '이용규', '강호동', '박승철', '김지은',  '강호동',  '강호동']
print(students)

students.remove('강호동')
print(students)

while '강호동' in students:
    students.remove('강호동')

print(students)
```

```python
#extend()
group1 = ['홍길동', '박찬호', '이용규']
group2 = ['강호동', '박승철', '김지은']
print('group1: {}'.format(group1))
print('group2: {}'.format(group2))

print()

group1.extend(group2)
print('group1: {}'.format(group1))
print('group2: {}'.format(group2))
```

```python
# +
group1 = ['홍길동', '박찬호', '이용규']
group2 = ['강호동', '박승철', '김지은']
print('group1: {}'.format(group1))
print('group2: {}'.format(group2))

print()

result = group1 + group2
print('group1: {}'.format(group1))
print('group2: {}'.format(group2))
print('result: {}'.format(result))
```

```python
#sort()
students = ['홍길동', '박찬호', '이용규', '강호동', '박승철', '김지은']
print('students: {}'.format(students))

students.sort() #ascending
print('students: {}'.format(students))


students.sort(reverse=True) #descending
print('students: {}'.format(students))

print()

numbers = [2, 50, 0.12, 1, 9, 7, 17, 35, 100, 3.14]
print('numbers: {}'.format(numbers))

numbers.sort()
print('numbers: {}'.format(numbers))
numbers.sort(reverse=True)
print('numbers: {}'.format(numbers))
```

```python
#reverse()
students = ['홍길동', '박찬호', '이용규', '강호동', '박승철', '김지은',  '강호동']
print('students: {}'.format(students))

students.reverse()
print('students: {}'.format(students))

print()

numbers = [2, 50, 0.12, 1, 9, 7, 17, 35, 100, 3.14]
print('numbers: {}'.format(numbers))

numbers.reverse()
print('numbers: {}'.format(numbers))
```

```python
# n:m 
#'n <= @ < m'
students = ['홍길동', '박찬호', '이용규', '강호동', '박승철', '김지은']
print('students: {}'.format(students))
print('students[2:4]: {}'.format(students[2:4]))
print('students[:4]: {}'.format(students[:4]))
print('students[2:]: {}'.format(students[2:]))
print('students[2:-2]: {}'.format(students[2:-2]))
print('students[-5:-2]: {}'.format(students[-5:-2]))

print()

numbers = [2, 50, 0.12, 1, 9, 7, 17, 35, 100, 3.14]
print('numbers: {}'.format(numbers))
print('numbers[2:4]: {}'.format(numbers[2:4]))
print('numbers[:4]: {}'.format(numbers[:4]))
print('numbers[2:]: {}'.format(numbers[2:]))
print('numbers[2:-2]: {}'.format(numbers[2:-2]))
print('numbers[-5:-2]: {}'.format(numbers[-5:-2]))

print()
print('type(numbers[-5:-2]): {}'.format(type(numbers[-5:-2])))
print('type(numbers[-5:-4]): {}'.format(type(numbers[-5:-4])))
print('numbers[-5:-4]): {}'.format(numbers[-5:-4]))

print()
students = ['홍길동', '박찬호', '이용규', '강호동', '박승철', '김지은']
print('students: {}'.format(students))


students[1:4] = ['park', 'lee', 'gang']
print('students: {}'.format(students))
```

```python
#slice([,])
students = ['홍길동', '박찬호', '이용규', '강호동', '박승철', '김지은']

print('students[slice(2, 4)]: {}'.format(students[slice(2, 4)]))
print('students[slice(4)]: {}'.format(students[slice(4)]))
print('students[slice(2, len(students)]: {}'.format(students[slice(2, len(students))]))
print('students[slice(len(students)-5, len(students)-2]: {}'.format(students[slice(len(students)-5, len(students)-2)]))
```
