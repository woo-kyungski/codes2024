```python
def cmTomm(n):
    return round(n * 10, 3)

def cmToinch(n):
    return round(n * 0.393, 3)

def cmTom(n):
    return round(n * 0.01, 3)

def cmToft(n):
    return round(n * 0.032, 3)

print(f'10cm :{cmTomm(10)}mm')
print(f'10cm :{cmToinch(10)}inch')
print(f'10cm :{cmTom(10)}m')
print(f'10cm :{cmToft(10)}ft')
```
