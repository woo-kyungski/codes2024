```python
#수학 관련 함수

#최댓값
listVar = [2, 5, 3,.14, 58, 10, 2]
print(f'sum(listVar): {max(listVar)}')

#최솟값
listVar = [2, 5, 3.14, 58, 10, 2]
print(f'sum(listVar): {min(listVar)}')

#거듭제곱
print(f'pow(13, 2): {pow(13, 4)}')

#반올림
print(f'round(3.141592, 2): {round(3.141592, 2)}')

import math

#절댓값
print(f'math.fabs(-10): {math.fabs(-10)}')
print(f'math.fabs(-12345): {math.fabs(-12345)}')

#올림
print(f'math.ceil(5.21): {math.ceil(5.21)}')

#내림
print(f'math.floor(5.21): {math.floor(5.21)}')

#버림
print(f'math.trunc(5.21): {math.trunc(5.21)}')

#최대공약수
print(f'math.gcd(14, 21) : {math.gcd(14, 21)}')

#팩토리얼
print(f'math.factorial(10) : {math.factorial(10)}')

#제곱근
print(f'math.sqrt(4) :{math.sqrt(4)}')
print(f'math.sqrt(12) :{math.sqrt(12)}')
```

```python
class Car:

    def __init__(self, col, len):
        self.color = col
        self.length = len

    def doStop(self):
        print('STOP')

    def doStart(self):
        print('START')

    def printCarinfo(self):
        print(f'self.color: {self.color}')
        print(f'self.length: {self.length}')
```

```python
car1 = Car('red', 200)
car2 = Car('blue', 300)
```

```python
car1.printCarinfo()
car2.printCarinfo()
```

```python
car1.doStop()
```

```python
car2.doStart()
```

```python
class NewGenerationPC:

    def __init__(self, name, cpu, memory, ssd):
        self.name = name
        self.cpu = cpu
        self.memory = memory
        self.ssd = ssd

    def doExcel(self):
        print('EXCEL RUN!')

    def doPhotoshop(self):
        print('PHOTOSHOP RUN!')

    def printPCINFO(self):
        print(f'self.name: {self.name}')
        print(f'self.cpu: {self.cpu}')
        print(f'self.memory: {self.memory}')
        print(f'self.ssd: {self.ssd}')
```

```python
myPC = NewGenerationPC('myPC', 'i5', '16G', '256G')
myPC.printPCINFO()
```

```python
friendPC = NewGenerationPC('friendPC', 'i7', '32G', '512')
friendPC.printPCINFO()
```

```python
myPC.cpu = 'i9'
myPC.memory = '64G'
myPC.ssd = '1T'
myPC.printPCINFO()
```

```python
class Robot:

    def __init__(self, color, height, weight):
        self.color = color
        self.height = height
        self.weight = weight

    def printRobotinfo(self):
        print(f'color: {self.color}')
        print(f'height: {self.height}')
        print(f'weight: {self.weight}')
        
```

```python
rb1 = Robot('red', 200, 80)
```

```python
rb2 = Robot('blue', 300, 120)
```

```python
rb3 = rb1 #얕은 복사: 주소값만 복사되어 할당됨
```

```python
rb1.printRobotinfo()
rb2.printRobotinfo()
rb3.printRobotinfo()
```

```python
rb1.color = 'gray'
rb1.height= 250
rb1.weight = 100
```

```python
rb1.printRobotinfo()
rb2.printRobotinfo()
rb3.printRobotinfo()
```
