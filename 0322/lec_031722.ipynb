```python
fx = np.linspace(100000, 700000, 100) #경향선을 그리기 위한 x 데이터 생성 / a부터 b까지 n개의 등간격 데이터 생성
fx 
```

```python
def drawfraph():

    plt.figure(figsize=(14, 10))
    plt.scatter(data_result['인구수'], data_result['소계'], s=50)
    plt.plot(fx, f1(fx), ls='--', lw=3, color='g')
    plt.xlabel('인구수')
    plt.ylabel('CCTV')
    plt.grid(True)
    plt.show()
    
drawfraph()
```

```python
fp1 = np.polyfit(data_result['인구수'], data_result['소계'], 1)
f1 = np.poly1d(fp1)
fx = np.linspace(100000, 700000, 100)
```

```python
data_result.head()
```

```python
data_result['오차'] = data_result['소계'] - f1(data_result['인구수'])
data_result.head()
```

```python
df_sort_f = data_result.sort_values(by='오차', ascending=False)
df_sort_t = data_result.sort_values(by='오차', ascending=True)
```

```python
df_sort_f.head()
```

```python
df_sort_t.head()
```

```python
from matplotlib.colors import ListedColormap

color_step = ['#e74c3c', '#2ecc71', '#95a9a6', '#2ecc71', '#3498db', '#3489db']
my_cmap = ListedColormap(color_step)
```

```python
def drawgraph():

    plt.figure(figsize=(14, 10))
    plt.scatter(data_result['인구수'], data_result['소계'], s=50, 
               c = data_result['오차'],
               cmap=my_cmap)
    plt.plot(fx, f1(fx), ls='--', lw=3, color='g')

    for n in range(5):
        plt.text(df_sort_f['인구수'][n]* 1.02, df_sort_f['소계'][n]* 0.98, df_sort_f.index[n], fontsize=10)
        plt.text(df_sort_t['인구수'][n]* 1.02, df_sort_t['소계'][n]* 0.98, df_sort_t.index[n], fontsize=10)
    
    plt.xlabel('인구수')
    plt.ylabel('CCTV')
    plt.colorbar()
    plt.grid(True)
    plt.show()
    
drawgraph()
```
