```python
#팩토리얼 재귀함수로 구현하기
def recursionFun(n):
    if n == 1:
        return n

    return n * recursionFun(n-1)

input_num = int(input())
print(recursionFun(input_num))
```

```python
def getPermutationCnt(n, r, logprint = True):

    result = 1
    for n in range(n, n-r -1):
        result *= n

    return result

from itertools import permutations as pt

numN = int(input('numN 입력: '))
numR = int(input('numR 입력: '))

print(f'{numN}P{numR} : {pt(str(getPermutationCnt(numN, numR)))}')
```
