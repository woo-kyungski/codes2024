```python
class Temcls:
    
    def __init__(self, n, s):
        self.number = n
        self.str = s

    def printClsInfo(self):
        print(f'self.number: {self.number}')
        print(f'self.str: {self.str}')

#얕은 복사
tc1 = Temcls(10, 'Hello')
tc2 = tc1

tc1.printClsInfo()
tc2.printClsInfo()

tc2.number = 3.14
tc2.str = 'Bye'

tc1.printClsInfo()
tc2.printClsInfo()
```

```python
#깊은 복사
import copy
tc1 = Temcls(10, 'Hello')
tc2 = copy.copy(tc1)

tc1.printClsInfo()
tc2.printClsInfo()

tc2.number = 3.14
tc2.str = 'Bye'

tc1.printClsInfo()
tc2.printClsInfo()
```

```python
scores = [9, 8, 5, 7, 6, 10]
scorecopy = []
scorescopy = scores

print(scores)
print(scorescopy)


scores = [9, 8, 5, 7, 6, 10]
scorecopy = []

for s in scores:
    scorecopy.append(s)

print(scores)
print(scorescopy)


scores = [9, 8, 5, 7, 6, 10]
scorecopy = []

scorecopy.extend(scores)
print(scores)
print(scorescopy)


scores = [9, 8, 5, 7, 6, 10]
scorecopy = []

scorecopy = scores.copy()
print(scores)
print(scorescopy)

scores = [9, 8, 5, 7, 6, 10]
scorecopy = []

scorecopy = scores[:]
print(scores)
print(scorescopy)
```

```python
#상속
class NormalCar:

    def drive(self):
        print('[NormalCar] drive() called!!')

    def back(self):
        print('[NormalCar] back() called!!')
```

```python
class TurboCar(NormalCar):

    def turbo(self):
        print('[TurboCar] turbo() called!!')
```

```python
myTurbocar = TurboCar()

myTurbocar.turbo()
myTurbocar.drive()
myTurbocar.back()
```

```python
class P_Class:
    def __init__(self, pNum1, pNum2):
        print('[P_Class] __init__() called!')
        self.pNum1 = pNum1
        self.pNum2 = pNum2


class C_Class(P_Class):
    
    def __init__(self, cNum1, cNum2):
        print('[C_Class] __init__() called!')
        
        # P_Class.__init__(self, cNum1, cNum2)
        super().__init__(cNum1, cNum2)
        self.pNum1 = cNum1
        self.pNum2 = cNum2

cls = C_Class(10, 20)    
```

```python
#오버라이드
class TriangleArea:

    def __init__(self, w, h):
        self.width = w
        self.height = h

    def printTriangleAreaInfo(self):
        print(f'self.width: {self.width}')
        print(f'self.height: {self.height}')

    def getArea(self):
        return self.width * self.height /2

class NewTriangleArea(TriangleArea):

    def __init__(self, w, h):
        super().__init__(w, h)

    def getArea(self):
        return str(super().getArea()) + 'mm'

ta = NewTriangleArea(7, 5)
ta.printTriangleAreaInfo()
triangleArea = ta.getArea()
print(f'triangleArea: {triangleArea}')
```

```python
from abc import ABCMeta
from abc import abstractclassmethod

class Airplane(metaclass=ABCMeta):

    @abstractclassmethod
    def flight(self):
        pass

    def forward(self):
        print('전진!')

    def backward(self):
        print('후진!')

class Airliner(Airplane):

    def __init__(self, c):
        self.color = c

    def flight(self):
        print('시속 400km/h 비행!')

al = Airliner('red')
al.flight()
al.forward()
al.backward()
```

```python
#예외처리

a = int(input('숫자를 입력해주세요: '))
b = int(input('숫자를 입력해주세요: '))

print(f'a + b: {a + b}')
print(f'a - b: {a - b}')
print(f'a * b: {a * b}')

try:
    print(f'a / b: {a / b}')
    
except:
    print('예상치 못한 예외가 발생했습니다.')
    print('다른 프로그램 실행에는 문제 없습니다.')
```

```python
#예외처리
nums = []

n = 1
while n < 6:

    try:
        num = int(input('input number: '))

    except:
        print('예외 발생!!')
        continue

    nums.append(num)
    n += 1

print(f'nums: {nums}')
```

```python
#예외처리2
nums = []
n = 1
while n < 6:

    try:
        num = int(input('input number: '))

    except:
        print('예외 발생!!')
        continue
    
    else:
        if num% 2 == 0:
            nums.append(num)
            n += 1

        else:
            print('입력된 숫자는 홀수 입니다.' , end='')
            print('다시 입력 하세요.')
            continue

print(f'nums: {nums}')
```
