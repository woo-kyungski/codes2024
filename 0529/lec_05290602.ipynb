```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings(action='ignore')
%matplotlib inline
```

```python
population = pd.read_excel('C:/Users/Administrator/Desktop/woo/codes2024/0529/data/07_population_raw_data.xlsx', header=1)
population.fillna(method='pad', inplace=True) #pad: na 값 이전 값으로 채우기 
population.head()
```

```python
population.info()
```

```python
population.rename(
    columns={'행정구역(동읍면)별(1)': '광역시도',
            '행정구역(동읍면)별(2)': '시도',
            '계': '인구수',
            }, inplace=True)

population.tail()
```

```python
population.head()
```

```python
population = population[population['시도'] !='소계']
population.head()
```

```python
population.is_copy = False

population.rename(
    columns={'항목': '구분'}, inplace=True)

population.head()
```

```python
population.loc[population['구분'] =='총인구수 (명)', '구분'] = '합계'
population.loc[population['구분'] =='남자인구수 (명)', '구분'] = '남자'
population.loc[population['구분'] =='여자인구수 (명)', '구분'] = '여자'
population.head()
```

```python
population['20-39세'] = (
    population['20 - 24세'] + population['25 - 29세'] + population['30 - 34세'] + population['35 - 39세'])

population['65세 이상'] = (
    population['65 - 69세'] + population['70 - 74세'] + population['75 - 79세'] + population['80 - 84세'] + population['85 - 89세'] + population['90 - 94세'] + population['95 - 99세'] + population['100+'])

population.head()
```

```python
pop = pd.pivot_table(
    data=population,
    index=['광역시도', '시도'],
    columns=['구분'],
    values=['인구수','20-39세', '65세 이상'])

pop
```

```python
pop['소멸비율'] = pop['20-39세', '여자'] / (pop['65세 이상', '합계'] / 2)
pop.tail()
```

```python
pop['소멸위기지역'] = pop['소멸비율'] < 1.0
pop
```

```python
pop[pop['소멸위기지역'] == True]
```

```python
pop[pop['소멸위기지역'] == True].index.get_level_values(1)
```

```python
pop.reset_index(inplace=True)
pop.head()
```

```python
tmp_columns = [
    pop.columns.get_level_values(0)[n] + pop.columns.get_level_values(1)[n]
    for n in range(0, len(pop.columns.get_level_values(0)))
]

tmp_columns
```

```python
pop.columns = tmp_columns
pop.head()
```

```python
tmp_gu_dict = {
    '수원': ['장안구', '권선구', '팔달구', '영통구'],
    '성남': ['수정구', '중원구', '분당구'],
    '안양': ['만안구', '동안구'],
    '안산': ['상록구', '단원구'],
    '고양': ['덕양구', '일산동구', '일산서구'],
    '용인': ['처인구', '기흥구', '수지구'],
    '청주': ['상당구', '서원구', '흥덕구', '청원구'],
    '천안': ['동남구', '서북구'],
    '전주': ['완산구', '덕진구'],
    '포항': ['남구', '북구'],
    '창원': ['의창구', '성산구', '진해구', '마산함포구', '마산회원구'],
    '부천': ['오정구', '원미구', '소사구']
}
    
```

```python
for idx, row in pop.iterrows():
    if row['광역시도'][-3:] not in ['광역시', '특별시', '자치시']:
        si_name[idx] = row['시도'][:-1]

    elif row['광역시도'] == '세종특별자치시':
        si_name[idx] = '세종'

    else:
        if len(row['시도']) == 2:
            si_name[idx] == row['광역시도'][:2] + ' ' + row['시도']
        else:
            si_name[idx] == row['광역시도'][:2] + ' ' + row['시도'][:-1]
        
```

```python
for idx, row in pop.iterrows():
    if row['광역시도'][-3:] not in ['광역시', '특별시', '자치시']:
        for keys, values in tmp_gu_dict.items():
            if row['시도'] in values:
                if len(row['시도']) == 2:
                    si_name[idx] = keys + ' ' + row['시도']
                elif row['시도'] in ['마산합포구', '마산회원구']:
                    si_name[idx] = keys + ' ' + row['시도'][2:-1]

                else:
                    si_name[idx] = keys + ' ' + row['시도'][:-1]
```

```python
si_name
```

```python
for idx, row in pop.iterrows():
    if row['광역시도'][-3:] not in ['광역시', '특별시', '자치시']:
        if row['시도'][:-1] == '고성' and row['광역시도'] == '강원도':
            si_name[idx] = '고성(강원)'
    
        elif row['시도'][:-1] == '고성' and row['광역시도'] == '경상남도':
            si_name[idx] = '고성(경남)'
```
