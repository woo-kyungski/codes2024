```python
import requests
from bs4 import BeautifulSoup

def get_hotel_price():
    url = "https://www.booking.com/hotel/kr/paito-7.ko.html"
    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")
    
    # Find the price element (you may need to inspect the page source to locate the correct element)
    price_element = soup.find("span", class_="bui-price-display__value")
    
    if price_element:
        return price_element.text
    else:
        return "Price information not available."

if __name__ == "__main__":
    hotel_price = get_hotel_price()
    print(f"파티오세븐 호텔의 현재 가격: {hotel_price}")
```
