```python
def divCalculator(n1, n2):

    if n2 != 0:
        print(f'{n1} / {n2} = {n1 / n2}')
    else: 
        raise Exception('0으로 나눌 수 없습니다.')
```

```python
n1 = int(input('input number1: '))
n2 = int(input('input number2: '))
divCalculator(n1, n2)
```

```python
try: 
    divCalculator(n1, n2)

except Exception as e:
    print(f'Exception: {e}')
```

```python
class Notusezeroexception(Exception):

    def __init__(self, n):
        super().__init__(f'{n}은 사용할 수 없습니다.')

    def divCalculator(n1, n2):

        if n2 == 0:
            raise Notusezeroexception(n2)

        else: 
            print(f'{n1} / {n2} = {n1 / n2}')
```

```python
n1 = int(input('input number1: '))
n2 = int(input('input number2: '))

try:
    divCalculator(n1, n2)
except Notusezeroexception as e:
    print(e)
```

```python
class PasswordlengthshortException(Exception):
    def __init__(self, str):
        super().__init__(f'{str}: 길이 5미만!!')

class PasswordlengthlongException(Exception):
    def __init__(self, str):
        super().__init__(f'{str}: 길이 10초과!!')

class PasswordwrongException(Exception):
    def __init__(self, str):
        super().__init__(f'{str}: 잘못된 비밀번호!!')

adminPw = input('input password: ')

try:
    if len(adminPw) <5:
        raise PasswordlengthshortException(adminPw)

    elif len(adminPw) > 10:
        raise PasswordlengthlongException(adminPw)

    elif adminPw != 'admin1234':
        raise PasswordwrongException(adminPw)

    elif adminPw == 'admin1234':
        print('환영합니다')

except PasswordlengthshortException as e1:
    print(e1)

except PasswordlengthlongException as e2:
    print(e2)

except PasswordwrongException as e3:
    print(e3)    
```

```python
uri = 'C:/pythontxt/'

def writhPrimenumber(n):
    file = open(uri + 'prime_numbers.txt', 'a')
    file.write(str(n))
    file.write('\n') #개행
    file.close()
```

```python
inputNumber = int(input('0보다 큰 정수 입력: '))
for number in range(2, (inputNumber + 1)):
    flag = True
    for n in range(2, number):
        if number % n == 0:
            flag = False
            break
    if flag:
        writhPrimenumber(number)
```

```python
scoreDic = {}
uri = 'C:/pythontxt/'
with open (uri + 'scores.txt', 'r') as f:
    line = f.readline()

    while line != '':
        tempList = line.split(':')
        scoreDic[tempList[0]] = int(tempList[1].strip('\n'))
        line = f.readline()

print(f'scoreDic: {scoreDic}')
```
