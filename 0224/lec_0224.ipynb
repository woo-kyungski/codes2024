```python
cars = ('그랜저', '소나타', '말리부', '카니발', '소렌토')

n = 0
flag = True
while flag:
    print(cars[n])
    n += 1

    if n == len(cars):
        flag = False
```

```python
scores = {'kor':88, 'eng':55, 'mat':85, 'sci':57, 'his':82}
print(f'scores: {scores}')

minScore = 60
fStr = 'F'

if scores['kor'] < minScore: scores['kor'] = fStr
if scores['eng'] < minScore: scores['eng'] = fStr
if scores['mat'] < minScore: scores['mat'] = fStr
if scores['sci'] < minScore: scores['sci'] = fStr
if scores['his'] < minScore: scores['his'] = fStr

print(f'scores: {scores}')
```

```python
scores = {'kor':88, 'eng':55, 'mat':85, 'sci':57, 'his':82}
print(f'scores.keys(): {scores.keys()}')
print(f'type(scores.keys()): {type(scores.keys())}')
ks = list(scores.keys())
print(ks)
print(type(ks))

print()

print(f'scores.values(): {scores.values()}')
print(f'type(scores.values()): {type(scores.values())}')
vs = list(scores.values())
print(vs)
print(type(vs))
```

```python
#del pop()

memInfo = {'이름': '홍길동',
           '메일': '길동지메일',
           '학년': '3',
           '취미': ['농구', '게임']}

print(memInfo)

del memInfo['이름']
print(memInfo)

returnvalue = memInfo.pop('취미')
print(f'memInfo: {memInfo}')
print(f'returnvalue: {returnvalue}')
```
