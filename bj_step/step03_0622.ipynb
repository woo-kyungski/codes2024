```python
t = int(input())

index_list = []
a_list = []
b_list = []
result_list = []

for i in range(t):
    a, b = map(int, input().split(' '))
    index_list.append(i + 1)
    a_list.append(a)
    b_list.append(b)
    result_list.append(a + b)

for i in range(len(index_list)):
    print('Case #' + str(index_list[i]) + ': ' + str(a_list[i]) + ' + ' + str(b_list[i]) + ' = ' + str(result_list[i]))
```

```python
result_list = []
a, b = map(int, input().split(' '))
result_list.append(a + b)

while a + b != 0:
    a, b = map(int, input().split(' '))
    result_list.append(a + b)

for i in range(len(result_list)-1):
    print(result_list[i])
```
