1	10807	개수 세기		82867	135273	62.218% https://www.acmicpc.net/problem/10807

```python
n = int(input())
nn = list(map(int, input().split(' ')))
nnn = int(input())
result = 0
for i in range(len(nn)):
    if nn[i] == nnn:
        result += 1
        
print(result)
    
```

2	10871	 X보다 작은 수		177793	325499	54.249% https://www.acmicpc.net/problem/10871

```python
n, x = map(int, input().split(' '))
ns = list(map(int, input().split(' ')))
result = []
for i in range(n):
    if ns[i] < x:
        result.append(ns[i])
        
result_result = ''
for i in range(len(result)):
    result_result += str(result[i]) + ' '
    
print(result_result)
```

5	10810	 공 넣기		44492	85658	52.500% https://www.acmicpc.net/problem/10810

```python
n, m = map(int, input().split(' '))

n_list = n * [0]
for a in range(m):
    i, j, k = map(int, input().split(' '))
    for b in range(i, j + 1):
        n_list[b-1] = k
        
result = ''
for i in range(len(n_list)):
    result += str(n_list[i]) + ' '
    
print(result)
```

  6	10813	 공 바꾸기		42798	67694	63.715% https://www.acmicpc.net/problem/10813

```python
n, m = map(int, input().split(' '))
balls = [z for z in range(1, n + 1)]
for a in range(m):
        i, j = map(int, input().split(' '))
        balls[i-1], balls[j-1] = balls[j-1], balls[i-1]

result = ''
for i in range(len(balls)):
    result += str(balls[i]) + ' '
    
print(result)  
```

7	5597	 과제 안 내신 분..?	다국어	66441	128669	51.702% https://www.acmicpc.net/problem/5597

```python
n = 28
students = [0] * 30
bad_students = []

for i in range(28):
    a = int(input())
    students[a - 1] += 1
    
for i in range(30):
    if students[i] == 0:
        bad_students.append(i + 1)
        
bad_students.sort()
print(bad_students[0])
print(bad_students[1])
```

10	1546	 평균	성공스페셜 저지	145973	288587	50.110% https://www.acmicpc.net/problem/1546

```python
n = int(input())
points = list(map(int, input().split(' ')))
new_points = []
max_p = max(points)
for i in range(len(points)):
    new_points.append((points[i] / max_p) * 100)

new_total = 0
for i in range(len(new_points)):
    new_total += new_points[i]
    
print(new_total / len(new_points))   
```
