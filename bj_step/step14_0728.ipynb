2	14425	 문자열 집합		30094	55145	53.839% https://www.acmicpc.net/problem/14425

3	7785	 회사에 있는 사람	다국어	24675	59731	40.562% https://www.acmicpc.net/problem/7785

```python
n = int(input())

working_list = []
for i in range(n):
    staff, which = map(str,input().split(' '))
    if which == 'enter':
        working_list.append(staff)
    
    else:
        working_list.pop(working_list.index(staff))

for i in range(len(working_list)):
    print(working_list[i])
```

4	1620	 나는야 포켓몬 마스터 이다솜		42599	117943	34.437% https://www.acmicpc.net/step/49

```python
#결과는 나오지만 시간 초과
n, m = map(int, input().split(' '))
pkm_dict = {}
dict_pkm = {}

for i in range(1, n + 1):
    name = input()
    pkm_dict[name] = i
    dict_pkm[i] = name
    
for i in range(1, m + 1):
    request = input()
    try:
        print(pkm_dict[request])
        
    except:
        print(dict_pkm[int(request)])
```

5	10816	 숫자 카드 2		60907	155353	37.759% https://www.acmicpc.net/problem/10816

```python
#시간 초과
n = int(input())
sand_cards = list(map(int, input().split(' ')))

m = int(input())
num_cards = list(map(int, input().split(' ')))

how_many = [0] * m
for i in range(m):
    for j in range(n):
        if num_cards[i] == sand_cards[j]:
            how_many[i] += 1
            
print(*how_many)    
```

6	1764	 듣보잡		49737	115114	41.454% https://www.acmicpc.net/problem/1764

```python
#교집합 이용
n, m = map(int, input().split(' '))

a = set()
for i in range(n):
    a.add(input())
    
b = set()
for j in range(m):
    b.add(input())
    
result = sorted(list(a & b))

print(len(result))

for i in result:
    print(i)
```
