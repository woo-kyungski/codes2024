```python
num_list = "0123456789ABCDEFGHIJKLMNOPQRSTUVWXYZ"

n, b = input().split()
answer = 0
for i, num in enumerate(n[::-1]):
    answer += int(b) ** i * num_list.index(num)
print(answer)
```

2	11005	 진법 변환 2		25279	54401	46.382% https://www.acmicpc.net/problem/11005

```python
#https://velog.io/@wlrhkd49/%EB%B0%B1%EC%A4%80-11005-%EC%A7%84%EB%B2%95-%EB%B3%80%ED%99%98-2-Python
n, b = map(int, input().split())
result = ''
nums = "0123456789ABCDEFGHIJKLMNOPQRSTUVWXYZ"

while n:
    result += str(nums[n % b])
    n //= b

print(result::-1)
```

 3	2720	 세탁소 사장 동혁	다국어	26903	38777	70.200% https://www.acmicpc.net/problem/2720

```python
t = int(input())
rests = [25, 10, 5, 1]
for i in range(t):
    hm_qu = 0; hm_dm = 0; hm_nck = 0; hm_pn = 0
    hms = [hm_qu, hm_dm, hm_nck, hm_pn]
    c = int(input())
    for j in range(len(rests)):
        hms[j] += (c // rests[j])
        c = (c % rests[j])
    for k in range(len(hms)):
        print(hms[k], end=' ')
```

4	2903	 중앙 이동 알고리즘	다국어	20420	28733	71.634% https://www.acmicpc.net/problem/2903

6	1193	 분수찾기		65089	127919	51.734% https://www.acmicpc.net/problem/1193

```python
x = int(input())
line = 1

while x > line:
    x -= line
    line += 1

# 짝수일경우
if line % 2 == 0:
    dson = x
    dmom = line - x + 1
# 홀수일경우
elif line % 2 == 1:
    dson = line - x + 1
    dmom = x

print(f'{dson}/{dmom}')
```
