1	5086	 배수와 약수	다국어	37842	60748	63.175% https://www.acmicpc.net/problem/5086

```python
a = 1; b = 1
while a != 0 and b != 0:
    a, b = map(int, input().split(' '))
    if a == 0:
         break
            
    elif b % a == 0:
        print('factor')
        
    elif a % b == 0:
        print('multiple')
        
    else:
        print('neither') 
```

2	2501	 약수 구하기		34958	70933	49.973% https://www.acmicpc.net/problem/2501

3	9506	 약수들의 합	다국어	24443	48498	51.257% https://www.acmicpc.net/problem/9506

```python
n = 1
while n != -1:   
    n = int(input())
    if n == -1:
        break
        
    else:
        n_babies = []
        perfection_comment = ''
        for i in range(1, n + 1):
            if n % i == 0:
                n_babies.append(i)
        if sum(n_babies) == 2 * n: 
            for j in range(len(n_babies) - 1):
                if j == len(n_babies) - 2: 
                    perfection_comment += str(n_babies[j])
                else:
                    perfection_comment += str(n_babies[j]) + ' + '
            print(f'{n}'  + ' = ' + f'{perfection_comment}')
        else:
             print(f'{n}'  + ' is NOT perfect.')

      
```

4	1978	 소수 찾기		102022	215279	47.238% https://www.acmicpc.net/problem/1978

```python
n = int(input())
special_list = []
a = 1
while len(special_list) != n:
    
    i_temp = []
    for i in range(1, a + 1):
        if a % i == 0:
            i_temp.append(i)
            
    if  len(i_temp) == 2:
        special_list.append(a)
            
    a += 1
        
print(*special_list, sep=' ')   
```

```python
n = int(input())
nums = list(map(int, input().split(' ')))
result = 0
for i in range(n):
    nums_temp = []
    for j in range(1, nums[i] + 1):
        if nums[i] % j == 0:
            nums_temp.append(j)
            
    if len(nums_temp) == 2:
        result += 1
            
print(result)
```

5	2581	 소수		61394	154395	39.507% https://www.acmicpc.net/problem/2581

```python
m = int(input())
n = int(input())
dec_list = []
for i in range(m, n + 1):
    temp = []
    for j in range(1, i + 1):
        if i % j == 0:
            temp.append(j)

    if len(temp) == 2:
        dec_list.append(i)

if sum(dec_list) != 0:
    print(sum(dec_list))
    print(dec_list[0])

else:
    print(-1)
```
