1	1934	 최소공배수		44708	82301	55.369% https://www.acmicpc.net/problem/1934

```python
t = int(input())
for i in range(t):
    a, b = map(int, input().split(' '))
    a_d = []; b_d = []
    for j in range(2, a + 1):        
        if a % j == 0:
            a_d.append(j)
            a //= j
                
    for k in range(2, b + 1):
        if b % k == 0:
            b_d.append(k)
            b //= k

    result = list(set(a_d + b_d))
    print_result = 1
    for l in range(len(result)):
        print_result *= result[l]
        
    print(print_result)        
```

3	1735	 분수 합		17212	37069	47.191% https://www.acmicpc.net/problem/1735

```python
import math
a, b = map(int, input().split(' '))
c, d = map(int, input().split(' '))

r_m = b * d
r_s = a * d + b * c
n = math.gcd(r_s, r_m)

result = []
result.append(r_s // n)
result.append(r_m // n)
print(*result)
```

4	2485	 가로수		10603	25997	41.854% https://www.acmicpc.net/problem/2485

```python
#https://velog.io/@aswe0409/%EB%B0%B1%EC%A4%80-2485%EA%B0%80%EB%A1%9C%EC%88%98-%ED%8C%8C%EC%9D%B4%EC%8D%AC
import math

def f_gcd(a, b):
    while b > 0:
        a, b = b, a % b
        
    return a

n = int(input())
count = 0

tlist = list(sorted(int(input()) for i in range(n)))

sub = []
for j in range(len(tlist) - 1):
    sub.append(tlist[j + 1] - tlist[j])
    
gcd = sub[0]
for k in range(1, len(sub)):
    gcd = f_gcd(gcd, sub[k])
    
for l in sub:
    count += l // gcd - 1
    
print(count)
```

```python
#https://develop247.tistory.com/339
# n보다 작거나 같은 소수들의 리스트를 반환하는 함수
def get_primes(n):
    # 0과 1은 소수가 아니므로 False로 초기화
    is_prime = [False, False] + [True] * (n-1)
    primes = []
    for i in range(2, n+1):
        if is_prime[i]:
            primes.append(i)
            # i의 배수들을 모두 소수가 아닌 것으로 표시
            for j in range(i*2, n+1, i):
                is_prime[j] = False
    return primes

# 입력받은 두 수의 범위를 저장
m, n = map(int, input().split())

# n보다 작거나 같은 소수들의 리스트를 구함
primes = get_primes(n)

# m부터 n까지의 수 중에서 소수인 수를 출력
for p in primes:
    if p >= m:
        print(p)
```
