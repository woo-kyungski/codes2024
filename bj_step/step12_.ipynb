1	2798	 블랙잭	다국어	93287	185931	48.887% https://www.acmicpc.net/problem/2798

```python
#f런타임 에러
n, m = map(int, input().split(' '))
cards = list(map(int, input().split(' ')))
cards.sort()
cards.reverse()
card_choice = []
for i in range(m):
    if len(card_choice) < 3:
        card_choice.append(cards[i])
        
    if sum(card_choice) > 21:
        card_choice.pop(0)
  
print(sum(card_choice))
```

```python
n, m = map(int, input().split(' '))
cards = list(map(int, input().split(' ')))
result = 0
for i in range(n):
    for j in range(i + 1, n):
        for k in range(j + 1, n):
            if cards[i] + cards[j] + cards[k] > m:
                continue
            else:
                result = max(result, cards[i] + cards[j] + cards[k])
                
print(result)
```

2	2231	 분해합	다국어	75202	162414	45.403% https://www.acmicpc.net/problem/2231

```python
n = int(input())

for i in range(1, n + 1):
    div_sum = 0
    if i !=  n:
        div_sum += i
        for j in range(len(str(i))):
            div_sum += int(str(i)[j]) 
        
        if div_sum == n:
            print(i)
            break
        
    else:
        print(0)
```

3	19532	 수학은 비대면강의입니다		19774	39114	51.160% https://www.acmicpc.net/problem/19532

```python
a, b, c, d, e, f = map(int, input().split(' '))

for x in range(-999, 1000):    
    for y in range(-999, 1000):
        if (a * x) + (b * y) == c and (d * x) + (e * y) == f:
            print(x, y, end=' ')        
```

4	1018	 체스판 다시 칠하기		63976	128568	49.856% https://www.acmicpc.net/problem/1018

```python
N, M = map(int, input().split())
original = []
count = []

for _ in range(N):
    original.append(input())

for a in range(N-7):
    for b in range(M-7):
        index1 = 0
        index2 = 0
        for i in range(a, a+8):
            for j in range(b, b+8):
                if (i+j) % 2 == 0:
                    if original[i][j] != 'W':
                        index1 += 1
                    if original[i][j] != 'B':
                        index2 += 1
                else:
                    if original[i][j] != 'B':
                        index1 += 1
                    if original[i][j] != 'W':
                        index2 += 1
        count.append(min(index1, index2))

print(min(count))
```
