```python
#while문으로 다시 짜 보자
t = int(input())
for i in range(t):
    a, b = map(int, input().split(' '))
    
    result = []
    for j in range(2, a + 1):
        if a % j == 0:
            result.append(j)
            a %= j

    for k in range(2, b + 1):
        if b % k == 0:
            result.append(k)
            b %  k

    result = set(result)
    result = list(result)
    result_result = 1
    for l in range(len(result)):
        result_result *= result[l]

    print(result_result)
```
