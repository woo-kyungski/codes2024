1	2738	 행렬 덧셈		47951	91130	53.183% https://www.acmicpc.net/problem/2738

```python
#터미널에서는 실행이 잘 되는데 제출만 하면 런타임 에러가 뜬다,,
n, m = map(int, input().split(' '))

a_a =[]; b_a =[] ; c_a =[]
aa = [a_a, b_a, c_a]
for i in range(n):
    a, b, c = map(int, input().split(' '))
    a_a.append(a)
    b_a.append(b)
    c_a.append(c)

a_b = []; b_b = []; c_b = []
bb = [a_b, b_b, c_b]
for i in range(n):
    a, b, c = map(int, input().split(' '))
    a_b.append(a)
    b_b.append(b)
    c_b.append(c)
    
a_r = [0, 0, 0]
b_r = [0, 0, 0]
c_r = [0, 0, 0]
rr = [a_r, b_r, c_r]

for i in range(n):
    for j in range(m):
        rr[i][j] += aa[i][j] + bb[i][j]

for i in range(len(rr)):
    if i > 0:
        print('')
    for j in range(len(rr)):
        print(rr[j][i], end=' ')
       
```

```python
N, M = map(int, input().split())

A, B = [], []

for _ in range(N):
    temp = list(map(int, input().split()))
    A.append(temp)

for _ in range(N):
    temp = list(map(int, input().split()))
    B.append(temp)
    
for i in range(N):
    for j in range(M):
        print(A[i][j] + B[i][j],end=' ')
    print()
```

2	2566	 최댓값	스페셜 저지	38498	96462	41.149% https://www.acmicpc.net/problem/2566

```python
max_num = 0
max_list_index = 0
max_num_index = 0

for i in range(9):
    num_list = list(map(int, input().split(' ')))
    if max(num_list) > max_num:
        max_num = max(num_list)
        max_list_index = i
        max_num_index = num_list.index(max_num)
        
print(max_num)
max_list_index += 1
max_num_index += 1
print(max_list_index, max_num_index)
```

3	10798	 세로읽기		37793	78731	48.547% https://www.acmicpc.net/problem/10798

```python
outputs = [''] * 15
result = ''

for i in range(5):
    letters = input()
    for j in range(len(letters)):
        try:
            outputs[j] += letters[j]
        except:
            pass
            
for i in range(len(outputs)):
    result += outputs[i]
    
print(result)
```

# 포기`

```python
# https://ji-gwang.tistory.com/325
N = int(input())
array = [[0]*100 for _ in range(100)]  # 도화지 범위 초기화
for _ in range(N):  # 입력 받은 도화지 개수만큼 돈다.
    y1, x1 = map(int, input().split())  # 왼쪽아래 x,y 좌표를 받는다.

    for i in range(x1, x1 + 10):  # 세로를 돈다.
        for j in range(y1, y1 + 10):  # 가로를 돈다.
            array[i][j] = 1  # 해당 범위 값을 0에서 1로 바꿔준다.

result = 0  # 넓이를 출력할 변수
for k in range(100):  # 전체 도화지를 돌면서
    result += array[k].count(1)  # 1 개수만 세어준다

print(result)
```
