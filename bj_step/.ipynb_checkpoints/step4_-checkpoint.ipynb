1	10807	개수 세기		82867	135273	62.218% https://www.acmicpc.net/problem/10807

```python
n = int(input())
nn = list(map(int, input().split(' ')))
nnn = int(input())
result = 0
for i in range(len(nn)):
    if nn[i] == nnn:
        result += 1
        
print(result)
    
```

2	10871	 X보다 작은 수		177793	325499	54.249% https://www.acmicpc.net/problem/10871

```python
n, x = map(int, input().split(' '))
ns = list(map(int, input().split(' ')))
result = []
for i in range(n):
    if ns[i] < x:
        result.append(ns[i])
        
result_result = ''
for i in range(len(result)):
    result_result += str(result[i]) + ' '
    
print(result_result)
```

3	10818	 최소, 최대		176686	392071	43.962% https://www.acmicpc.net/problem/10818

4	2562	 최댓값		153033	332355	45.444% https://www.acmicpc.net/problem/2562

```python
n_list = []
for i in range(9):
    n_list.append(int(input()))
    
print(max(n_list))

max_index = 0
for i in range(9):
    if n_list[i] == (max(n_list)):
        max_index = i + 1
        
print(max_index)
```

5	10810	 공 넣기		44492	85658	52.500% https://www.acmicpc.net/problem/10810

```python
n, m = map(int, input().split(' '))

n_list = n * [0]
for a in range(m):
    i, j, k = map(int, input().split(' '))
    for b in range(i, j + 1):
        n_list[b-1] = k
        
result = ''
for i in range(len(n_list)):
    result += str(n_list[i]) + ' '
    
print(result)
```
