6	10809	 알파벳 찾기		123794	230787	53.328% https://www.acmicpc.net/problem/10809

```python
s = input()
alphabets = ['a', 'b', 'c', 'd', 'e',
             'f', 'g', 'h', 'i', 'j',
             'k', 'l', 'm', 'n', 'o',
             'p', 'q', 'r', 's', 't',
             'u', 'v', 'w', 'x', 'y',
             'z']

result = [0] * len(alphabets)

for i in range(len(alphabets)):
    if alphabets[i] in s:
        result[i] += s.find(alphabets[i])
        
        
    else: 
        result[i] -= 1

for i in range(len(result)):
    print(result[i], end=' ')
```
