4	10988	 팰린드롬인지 확인하기		49979	82135	61.142% https://www.acmicpc.net/problem/10988

```python
word_f = input()
word_f_list = [word_f[a] for a in range(len(word_f))]
word_f_list.reverse()
f_word = ''
for i in range(len(word_f_list)):
    f_word += word_f_list[i]

print(1) if f_word == word_f else print(0)
```

5	1157	 단어 공부		120008	299828	39.774% https://www.acmicpc.net/problem/1157

```python
the_word = input()
letter_particles = [the_word[i] for i in range(len(the_word))]
for i in range(len(letter_particles)):
    if ord(letter_particles[i]) <= 90:
        letter_particles[i] = chr(ord(letter_particles[i]) + 32)
      
letter_list = set(letter_particles)
letter_list = list(letter_list)
letter_freq = [0] * len(letter_list)

for i in range(len(letter_particles)):
    for j in range(len(letter_list)):
        if letter_particles[i] == letter_list[j]:
            letter_freq[j] += 1

max_freq = []
for i in range(len(letter_freq)):
    if letter_freq[i] == max(letter_freq):
        max_freq.append(letter_list[i])

result = ''
if len(max_freq) >= 2:
    print('?')

else: 
    for i in range(len(letter_freq)):
        if letter_freq[i] == max(letter_freq):
            result = chr(ord(letter_list[i]) - 32)

print(result)
```

7	1316	 그룹 단어 체커		88144	165881	53.492% https://www.acmicpc.net/problem/1316

```python
n = int(input())
group_count = n
for i in range(n):
    g_als= []
    g_word = input()
    for j in range(len(g_word)):
        if j == 0:
            g_als.append(g_word[j])
        else:
            if g_word[j] == g_word[j-1]:
                continue
                
            else:
                if g_word[j] in g_als:
                    group_count -= 1
                    break
                    
                else:
                    g_als.append(g_word[j])

print(group_count)
```

8	25206	 너의 평점은	스페셜 저지	27037	61864	44.067% https://www.acmicpc.net/problem/25206

```python
grade = ['A+', 	'A0', 'B+', 'B0', 'C+', 'C0', 'D+', 'D0', 'F']	
points = [4.5, 4.0, 3.5, 3.0, 2.5, 2.0, 1.5, 1.0, 0.0]
chi_points = []
chi_grades = []
devision_mom = 0
for i in range(20):
    chi_point = input()
    if chi_point[-1] == 'P':
        pass
    else:
        devision_mom += float(chi_point[-6:-2])
        for j in range(len(grade)):
            if chi_point[-2:] == grade[j]:
                chi_points.append(float(chi_point[-7:-2]) * points[j])

                
total = 0
for j in range(len(chi_points)):
    total += chi_points[j]
    
print(total / devision_mom) 
```
