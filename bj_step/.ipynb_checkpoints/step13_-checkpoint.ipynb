```python
n = int(input())
n_list = []
for i in range(n):
    num = int(input())
    n_list.append(num)
    
n_list.sort()
for i in range(len(n_list)):
    print(n_list[i])    
```

5	10989	 수 정렬하기 3	언어 제한	72417	304112	23.792% https://www.acmicpc.net/problem/10989

```python
#메모리 초과
n = int(input())
num_list = []
for i in range(n):
    num = int(input())
    num_list.append(num)
    
num_list.sort()
num_list.reverse()

for j in range(n):
    print(num_list[j])
```

```python
#https://kill-xxx.tistory.com/entry/Python-%EB%B0%B1%EC%A4%80-10989%EB%B2%88-%EC%88%98-%EC%A0%95%EB%A0%AC%ED%95%98%EA%B8%B0-3-%EB%A9%94%EB%AA%A8%EB%A6%AC-%EC%B4%88%EA%B3%BC-%EC%98%A4%EB%A5%98-%ED%95%B4%EA%B2%B0

import sys
input = sys.stdin.readline

#계수정렬 활용
n = int(input())
arr = [0] * (10000 + 1) # 입력값이 10000개까지 주어지니 10000 + 1개의 리스트 선언

#각 입력값에 해당하는 인덱스의 값 증가
for _ in range(n):
    arr[int(input())] += 1
  
#arr에 기록된 정보 확인
for i in range(len(arr)):
    if arr[i] != 0: #0이 아닌 데이터, 즉 입력받은 데이터들을 출력
        for _ in range(arr[i]):
            print(i)
```

6	1427	 소트인사이드		65880	100723	65.725% https://www.acmicpc.net/problem/1427

```python
n = int(input())
n = str(n)
nn = []
for i in range(len(n)):
    nn.append(n[i])
nn.sort()
nn.reverse()
result = ''
for i in range(len(nn)):
    result += nn[i]
    
print(result)
```

7	11650	 좌표 정렬하기		71765	148963	48.193% https://www.acmicpc.net/problem/11650

```python
#시간 초과
n = int(input())
coods = []
x_list = []
y_list = []

for i in range(n):
    cood = list(map(int, input().split(' ')))
    coods.append(cood)
    x_list.append(cood[0])
    y_list.append(cood[1])
    
x_list.sort()
y_list.sort()

results = []  

for i in range(n):
    for j in range(n):
        if x_list[i] == coods[j][0]:
            results.append(str(coods[j][0])+ ' '+ str(coods[j][1]))

results = set(results)
results = list(results)
results.sort()
print(results, end=' ')
```

8	11651	 좌표 정렬하기 2		54632	83650	66.881% https://www.acmicpc.net/problem/11651

```python
n = int(input())
coods = []
for i  in range(n):
    cood = list(map(int, input().split(' ')))
    coods.append(cood)
    
coods.sort(key = lambda x: (x[1], x[0]))
for j in coods:
    print(j[0], j[1])
```

9	1181	 단어 정렬		82267	195549	40.429% https://www.acmicpc.net/problem/1181

```python
n = int(input())
word_list = []
len_list = []
for i in range(n):
    word = input()
    word_list.append(word)
    len_list.append(len(word))
 
word_list = set(word_list)
word_list = list(word_list)
word_list.sort()
for j in range(1, max(len_list) + 1):
    for k in range(len(word_list)):
        if len(word_list[k]) == j:
            print(word_list[k])
```

10	10814	 나이순 정렬		68368	151499	43.686% https://www.acmicpc.net/problem/10814

```python
n = int(input())
guests = []
ages = []
for i in range(n):
    guest = input()
    guests.append(guest)
    ages.append(int(guests[i][:2]))
    
for j in range(1, max(ages) + 1):
    for k in range(len(guests)):
        if guests[k][:2] == j:
            print(guests[k])
```
