6	2525	 오븐 시계		114232	298274	38.158% https://www.acmicpc.net/problem/2525

```python
h, m = map(int, input().split(' '))
takes = int(input())

if takes + m >= 60:
    h += ((takes + m) // 60)
    m = (takes + m) % 60 

else:
    m += takes

if h >= 24:
    h = h % 24
print(h, m, sep=' ')
```

7	2480	 주사위 세개		111975	239526	46.795% https://www.acmicpc.net/problem/2480

```python
a, b, c = map(int, input().split(' '))

if a == b == c:
    print(10000 + 1000 * a)

elif a == b or a == c:
    print(1000 + 100 * a)

elif b == c:
    print(1000 + 100 * b)

else:
    print(max([a, b, c]) * 100)
   
```
