```python
stack = []
for i in range(int(input())):
    order_list = list(map(int, input().split(' ')))
    if order_list[0] == 1:
        stack.append(order_list[1])
        
    elif order_list[0] == 2:
        if stack:
            print(stack.pop())
        else:
            print(-1)
        
    elif order_list[0] == 3:
        print(len(stack))
        
    elif order_list[0] == 4:
        if len(stack) == 0:
            print(1)
        else:
            print(0)
        
    elif order_list[0] == 5:
        if len(stack) != 0:
            print(stack[-1])
        else:
            print(-1)
```

```python
#틀림
t = int(input())
for i in range(t):
    gwals = input()
    gwal_whether = 0
    for j in range(len(gwals)):
        if gwals[j] == '(':
            gwal_whether += 1
        else:
            gwal_whether -=1
    if gwal_whether == 0:
        print('YES')
        
    else:
        print('NO')          
```
