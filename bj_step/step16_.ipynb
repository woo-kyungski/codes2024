4	4949	 균형잡힌 세상	다국어	49372	145432	32.752% https://www.acmicpc.net/problem/4949

```python
#틀림
sentence = ''

while sentence != '.':
    sogwal = []
    daegwal = []
    sentence = input()
    if sentence == '.':
        break
        
    else:
        for i in sentence:
            if i == '(':
                sogwal.append(i)
            elif i == ')':
                if sogwal:
                    sogwal.pop()
                else:
                    print('no')
                    break

            elif i == '[':
                daegwal.append(i)

            elif i == ']':
                if daegwal:
                    daegwal.pop()
                else:
                    print('no')
                    break

        else:
            if len(sogwal) == 0 and len(daegwal) == 0:
                print('yes')

            else:
                print('no')    
```

```python
while True :
    a = input()
    stack = []

    if a == "." :
        break

    for i in a :
        if i == '[' or i == '(' :
            stack.append(i)
        elif i == ']' :
            if len(stack) != 0 and stack[-1] == '[' :
                stack.pop() # 맞으면 지워서 stack을 비워줌 0 = yes
            else : 
                stack.append(']')
                break
        elif i == ')' :
            if len(stack) != 0 and stack[-1] == '(' :
                stack.pop()
            else :
                stack.append(')')
                break
    if len(stack) == 0 :
        print('yes')
    else :
        print('no')
```
