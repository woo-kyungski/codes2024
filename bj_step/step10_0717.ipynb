5	9063	 대지		14916	23115	64.300% https://www.acmicpc.net/problem/9063

```python
n = int(input())
x_list = []; y_list = []
for i in range(n):
    x, y = map(int, input().split(' '))
    x_list.append(x)
    y_list.append(y)
    
print((max(x_list) - min(x_list)) * (max(y_list) - min(y_list)))
```

6	10101	 삼각형 외우기	다국어	25174	43958	57.392% https://www.acmicpc.net/problem/10101

```python
a = int(input())
b = int(input())
c = int(input())

if a + b + c == 180:
    if a == b == c == 60:
        print('Equilateral')

    elif a == b or b == c or a == c:
        print('Isosceles')

    else:
        print('Scalene')
else:
    print('Error')
```
