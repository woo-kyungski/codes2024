```python
import requests
from bs4 import BeautifulSoup

# 아고다 호텔 페이지 URL
url = "https://www.agoda.com/hotel-anteroom-seoul/hotel/seoul-kr.html?finalPriceView=1&isShowMobileAppPrice=false&cid=-1&numberOfBedrooms=&familyMode=false&adults=2&children=0&rooms=1&maxRooms=0&checkIn=2024-03-23&isCalendarCallout=false&childAges=&numberOfGuest=0&missingChildAges=false&travellerType=1&showReviewSubmissionEntry=false&currencyCode=KRW&isFreeOccSearch=false&isCityHaveAsq=false&los=1&searchrequestid=1f9f7184-6e06-4476-83dc-5c5ebb7d7738&ds=mZKPNUBAjtUdNXAe"

# User-Agent 설정 (필요한 경우)
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/107.0.0.0 Safari/537.36"
}

# 페이지 요청
response = requests.get(url, headers=headers)

# HTML 파싱
soup = BeautifulSoup(response.content, "html.parser")

# 호텔 이름 추출
hotel_name = soup.find("h3", class_="hotel-name").text.strip()
print(f"Hotel Name: {hotel_name}")

# 가격 추출 (예시)
price_element = soup.find("span", class_="price")
if price_element:
    price = price_element.text.strip()
    print(f"Price: {price}")
else:
    print("Price information not found.")

# 필요한 데이터를 추출하여 원하는 방식으로 활용하세요!
```
