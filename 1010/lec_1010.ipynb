```python
import pandas as pd

titanic_url = 'https://raw.githubusercontent.com/PinkWink/ML_tutorial/refs/heads/master/dataset/titanic.xls'
titanic = pd.read_excel(titanic_url)
titanic.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic['survived']
```

```python
titanic['survived'].value_counts()
```

```python
titanic['survived'].value_counts().plot.pie()
```

```python
titanic['survived'].value_counts().plot.pie(autopct='%1.1f%%')
```

```python
titanic['survived'].value_counts().plot.pie(autopct='%1.1f%%', shadow=True)
```

```python
titanic['survived'].value_counts().plot.pie(autopct='%1.1f%%', shadow=True, explode=[0, 0.05])
```

```python
f, ax = plt.subplots(1, 2, figsize=(16, 8))

titanic['survived'].value_counts().plot.pie(ax=ax[0], autopct='%1.1f%%', shadow=True, explode=[0, 0.05])
ax[0].set_title('Pie plot - survived')
ax[0].set_ylabel('')

sns.countplot(x='survived', data=titanic, ax=ax[1])
ax[1].set_title('Count plot - survived')

plt.show()
```
