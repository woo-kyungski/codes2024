```python
def calculator(n1, n2):
    result = n1 + n2
    
    return result #함수 호출부로 실행 결과를 반환

returnValue = calculator(20, 10)
print('returnValue: {}'.format(returnValue))
```

```python
totalVisit = 0

def countTotalVisit():
    global totalVisit

    totalVisit += 1
    print('누적방문객: {}'.format(totalVisit))

countTotalVisit()
countTotalVisit()
countTotalVisit()
```

```python
caculator = lambda n1, n2 : n1 + n2

returnValue = calculator(10, 20)
print('returnValue: {}'.format(returnValue))
```
