```python
import pandas as pd
import numpy as np

crime_anal_station = pd.read_csv('./data/02. crime_in_seoul_raw.csv', index_col=0, encoding='utf-8') #index_col '구분'을 인덱스 컬럼으로 설정
crime_anal_station.head()
```

```python
crime_anal_gu = pd.pivot_table(crime_anal_station, index='구별', aggfunc=np.sum)

del crime_anal_gu['lat']
crime_anal_gu.drop('lng', axis=1, inplace=True)

crime_anal_gu.head()
```

```python
crime_anal_gu['강도검거'] / crime_anal_gu['강도발생'] #강도검거율
```

```python
crime_anal_gu[['강도검거', '살인검거']].div(crime_anal_gu['강도발생'], axis=0).head()
```

```python
num = ['강간검거', '강도검거', '살인검거', '절도검거', '폭력검거']
den = ['강간발생', '강도발생', '살인발생', '절도발생', '폭력발생']

crime_anal_gu[num].div(crime_anal_gu[den].values).head()
```

```python
target = ['강간검거율', '강도검거율', '살인검거율', '절도검거율', '폭력검거율']
num = ['강간검거', '강도검거', '살인검거', '절도검거', '폭력검거']
den = ['강간발생', '강도발생', '살인발생', '절도발생', '폭력발생']

crime_anal_gu[target] = crime_anal_gu[num].div(crime_anal_gu[den].values) * 100
crime_anal_gu.head()
```

```python
del crime_anal_gu['강간검거']
del crime_anal_gu['강도검거']
crime_anal_gu.drop(['살인검거', '절도검거', '폭력검거'], axis=1, inplace=True)
crime_anal_gu.head()
```

```python
crime_anal_gu[crime_anal_gu[target] > 100] = 100
crime_anal_gu.head()
```

```python
crime_anal_gu.rename(columns={"강간발생": "강간",
                              "강도발생": "강도",
                              "살인발생": "살인", 
                              "절도발생": "절도",
                              "폭력발생": "폭력"}, inplace=True)

crime_anal_gu.head()
```

```python
crime_anal_gu.head()
```

```python
crime_anal_gu['강도'] / crime_anal_gu['강도'].max() #정규화
```

```python
col = ['살인', '강도', '강간', '절도', '폭력' ]
crime_anal_norm = crime_anal_gu[col] / crime_anal_gu[col].max()
crime_anal_norm.head()
```

```python
crime_anal_gu.head()
```

```python
anscombe = sns.load_dataset('anscombe')
anscombe.tail()
```

```python
anscombe['dataset'].unique()
```

```python
sns.set_style('darkgrid')
sns.lmplot(x='x',
          y='y',
          data=anscombe.query("dataset == 'I'"),
          ci=None,
          height=7)

plt.show()
```

```python
sns.set_style('darkgrid')
sns.lmplot(x='x',
          y='y',
          data=anscombe.query("dataset == 'I'"),
          ci=None, #신뢰구간
          height=7,
          scatter_kws={"s":50}) #점의 크기

plt.show()
```

```python
sns.set_style('darkgrid')
sns.lmplot(x='x',
          y='y',
          data=anscombe.query("dataset == 'I'"),
           order=1,
          ci=None, #신뢰구간
          height=7,
          scatter_kws={"s":50}) #점의 크기

plt.show()
```

```python
sns.set_style('darkgrid')
sns.lmplot(x='x',
          y='y',
          data=anscombe.query("dataset == 'I'"),
           order=2,
          ci=None, #신뢰구간
          height=7,
          scatter_kws={"s":50}) #점의 크기

plt.show()
```

```python
sns.set_style('darkgrid')
sns.lmplot(x='x',
          y='y',
          data=anscombe.query("dataset == 'I'"),
           robust=True,
          ci=None, #신뢰구간
          height=7,
          scatter_kws={"s":50}) #점의 크기

plt.show()
```
