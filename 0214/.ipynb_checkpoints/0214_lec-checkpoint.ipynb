```python
#등차수열

aN = [2, 5, 8, 11, 14, 17, 20, 20, 26, 29]
```

```python
input_a1 = int(input('a1 입력: '))
input_D = int(input('D 입력: '))
input_N = int(input('N 입력: '))

value_n = 0

n = 1
while n <= input_N:
    
    if n == 1:
        value_n = input_a1
        print('{}번째 항의 값: {}'.format(n, value_n))
        n += 1
        continue

    else:
        value_n += input_D
        print('{}번째 항의 값: {}'.format(n, value_n))
        n += 1

sum_n = input_N * (input_a1 + value_n) / 2

print('{}번째 항의 값: {}'.format(input_N, value_n))
print('{}번째 항까지의 Sum값: {}'.format(input_N, int(sum_n)))
```

```python
#등비수열
input_a1 = int(input('a1 입력: '))
input_R = int(input('R 입력: '))
input_N = int(input('N 입력: '))

value_N = 0
n = 1
while n <= input_N:

    if n == 1:
        value_N = input_a1
        print('{}번째 항의 값: {}'.format(n, value_N))
        n += 1
        continue

    value_N *= input_R
    print('{}번째 항의 값: {}'.format(n, value_N))
    n += 1

sum_N = input_a1 * (1 - (input_R ** input_N)) / (1 - input_R)
print('{}번째 항까지의 Sum값: {}'.format(input_N, int(sum_N)))
print('{}번째 항의 값: {}'.format(input_N, value_N))
```

```python
#계차수열
input_a1 = int(input('a1 입력: '))
input_an = int(input('an 입력: '))

input_b1 = int(input('b1 입력: '))
input_bn = int(input('bn 입력: '))

value_an = 0
value_bn = 0

n = 1
while n <= input_an:
    if n ==1:
        value_an = input_a1
        value_bn = input_b1
        print('an의 {}번째 항의 값: {}'.format(n, value_an))
        print('bn의 {}번째 항의 값: {}'.format(n, value_bn))
        n += 1
        continue

    value_an += value_bn
    value_bn += input_bn
    print('an의 {}번째 항의 값: {}'.format(n, value_an))
    print('bn의 {}번째 항의 값: {}'.format(n, value_bn))
    n += 1

print('')
print('an의 {}번째 항의 값: {}'.format(input_an, value_an))
print('bn의 {}번째 항의 값: {}'.format(input_an, value_bn))   
```
