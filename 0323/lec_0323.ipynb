```python
df = pd.read_excel('./data/02. sales-funnel.xlsx')
df.head()
```

```python
pd.pivot_table(df, index='Name')
```

```python
cols = ["Account", "Quantity", "Price"]

pd.pivot_table(df, index=["Name"], values=cols)
```

```python
df.pivot_table(index=['Name'], values=cols)
```

```python
df.pivot_table(index=['Name', 'Rep', 'Manager'], values=cols)
```

```python
df.pivot_table(index=['Manager', 'Rep'], values=cols)
```

```python
df.pivot_table(index=['Manager', 'Rep'], values='Price')
```

```python
df.pivot_table(index=['Manager', 'Rep'], values='Price', aggfunc=np.sum)
```

```python
df.pivot_table(index=['Manager', 'Rep'], values='Price', aggfunc=[np.sum, len])
```

```python
df.pivot_table(index=['Manager', 'Rep'], 
               values='Price',
               columns='Product',
               aggfunc=np.sum)
```

```python
#NaN 값 설정 : fill_value
df.pivot_table(index=['Manager', 'Rep'], 
               values='Price',
               columns='Product',
               aggfunc=np.sum,
              fill_value=0)
```

```python
df.pivot_table(index=['Manager', 'Rep', 'Product'], 
               values=['Price', 'Quantity'],
               aggfunc=np.sum,
              fill_value=0)
```

```python
df.pivot_table(index=['Manager', 'Rep', 'Product'], 
               values=['Price', 'Quantity'],
               aggfunc=[np.sum,np.mean],
              fill_value=0,
              margins=True) #총합
```
