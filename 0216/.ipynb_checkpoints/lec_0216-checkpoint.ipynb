```python
import random

rNum = random.randint(100, 1000)
print(f'rNum: {rNum}')

soinsu_list = []

n = 2
while n <= rNum:
    if rNum % n == 0:
        print(f'소인수: {n}')
        soinsu_list.append(n)
        rNum /= n

    else: 
        n += 1

print(f'soinsu_list: {soinsu_list}')

temp_num = 0
for s in soinsu_list:
    if temp_num != s:
        print(f'{s}\'s count: {soinsu_list.count(s)}')
        temp_num = s
```
