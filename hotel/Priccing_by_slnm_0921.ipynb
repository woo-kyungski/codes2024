```python
url = 'https://www.agoda.com/search?cid=1587497&currency=KRW&checkin=2024-09-21&checkout=2024-09-22&numberofchildren=0&mcid=27989&masterroomid=85512679&mpt=cXVFUHpxZUEyUDg3QnZuVVFmNW42Y1JOUDFZNGY3bllUNUsxNy9DWWYzNlk&pid=21b25fbf-c4b0-4644-8252-38a1e45d0fe6&audid=&audm=&mbl=L&baserate=174815.00&taxandfee=17481.00&basepricedisplay=B&displaycurrency=KRW&ccallout=FALSE&source=meta&hl=en&tag=hid276095%2Cpid21b25fbf-c4b0-4644-8252-38a1e45d0fe6&naver=1&n=65186&adults=2&rooms=1&los=1&selectedproperty=276095&city=14690&pslc=1&ds=OrFVUaKYwa%2B7jWgC'
```

```python
driver = webdriver.Chrome("C://chromedriver.exe") # webdriver 실행(크롬 창 생성)
driver.get(url) # url로 접속
```

```python
# 특정 클래스 혹은 id가 포함된 값들 중 하나만 가져온다 : find_element_by_css_selector
# 특정 클래스 혹은 id가 포함된 값들 모두 가져온다 : find_element_by_css_selector
# 클래스 표시 : . / id 표시 : #
# 예) h2.hp__hotel-name 해석 : h2 태그의 'hp__hotel-name' 클래스

price_b = driver.find_element_by_css_selector('div.Box-sc-kv6pi1-0.bWGdbw.PropertyCardPrice')
price = price_b.text
print(price)
```
